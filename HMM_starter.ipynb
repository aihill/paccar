```python
test_snapshots = pandas.read_pickle('test_snapshots.pkl')
train_snapshots = pandas.read_pickle('train_snapshots.pkl')
test = pandas.read_pickle('test.pkl')
train = pandas.read_pickle('train.pkl')

ok_test_snapshots = pandas.read_pickle('ok_test_snapshots.pkl')
ok_train_snapshots = pandas.read_pickle('ok_train_snapshots.pkl')
ok_test = pandas.read_pickle('ok_test.pkl')
ok_train = pandas.read_pickle('ok_train.pkl')
```

```python
category_var = ['Event Type Description','Brake Switch','Clutch Switch','Cruise Status','Dpf Regen Inhibit Sw', 
                'Dpf Thermal Mngmnt','Eng Coolant Level','DTCID',
                'Model Vehicle','ATA3','ATA6','ATA9','Repair Cost']

# drop identifier, timestamp, category and state columns before unsupervised learning
Xnum = train_snapshots.drop(['Veh Ref ID', 'Event DateTime', 'Repair Date', 'state'],axis = 1)

# drop categorical variables
Xnum = Xnum.drop(category_var,axis = 1)

# fill in X and lengths for passing into HMM
lengths = []
X = []
for i in train.index:
    veh_id = train.loc[i,'Chassis\nReference\nNumber']
    repairDate = train.loc[i,'Rpr_Dt']
    v_snapshots = Xnum[(train_snapshots['Veh Ref ID'] == veh_id) & (train_snapshots['Repair Date'] == repairDate)]
    
    if (len(v_snapshots) > 0):
        lengths.append(len(v_snapshots))
        #X = pandas.concat([X, v_snapshots])
        if(len(X) == 0):
            X = v_snapshots
        else:
            X = X.append(v_snapshots)
        

[sum(lengths),len(X)]
```

```python
num_states = 3
# HMM model with Gaussian emissions
model = hmm.GaussianHMM(n_components=num_states, covariance_type="diag", n_iter=10)
model.fit(X, lengths)


print(model.transmat_)


#print(model.predict(X,lengths))
```
